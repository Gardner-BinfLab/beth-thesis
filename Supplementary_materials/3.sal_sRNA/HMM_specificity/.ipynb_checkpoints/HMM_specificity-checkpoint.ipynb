## Boxplot of model sensitivity

First, count number of annotations per model per genome. These files have been filtered for annotations >70% model length.

```python
%bash
cat startingSeqs.gff | awk '{print $3,$1}' | sort | uniq -c| awk '{print $2","$1}' | sort -k2,2n > ~/Dropbox/PhD/sRNAs/model_stats.csv
cat trainingSeqs.gff | awk '{print $3,$1}' | sort | uniq -c| awk '{print $2","$1}' | sort -k2,2n >> ~/Dropbox/PhD/sRNAs/model_stats.csv
```

Make boxplot of ranges of number of annotations per genome for each model

```python
%%bash
#split into chunks for plotting (don't know how to easily do this in pandas yet)
cut -d ',' -f1 model_stats.csv | sort | uniq | head -n 50 > tmp1
for i in `cat tmp1`; do grep -E "$i," model_stats.csv ; done > model_stats_1.csv
cut -d ',' -f1 model_stats.csv | sort | uniq | head -n 100 | tail -n 50 > tmp2
for i in `cat tmp2`; do grep -E "$i," model_stats.csv ; done > model_stats_2.csv
cut -d ',' -f1 model_stats.csv | sort | uniq | head -n 150 | tail -n 50 > tmp3
for i in `cat tmp3`; do grep -E "$i," model_stats.csv ; done > model_stats_3.csv
cut -d ',' -f1 model_stats.csv | sort | uniq | head -n 200 | tail -n 50 > tmp4
for i in `cat tmp4`; do grep -E "$i," model_stats.csv ; done > model_stats_4.csv
cut -d ',' -f1 model_stats.csv | sort | uniq | head -n 250 | tail -n 50 > tmp5
for i in `cat tmp5`; do grep -E "$i," model_stats.csv ; done > model_stats_5.csv
cut -d ',' -f1 model_stats.csv | sort | uniq | tail -n 30 > tmp6
for i in `cat tmp6`; do grep -E "$i," model_stats.csv ; done > model_stats_6.csv

```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("model_stats.csv",header=None)
df.columns=['sRNA','count']
df1 = pd.read_csv("model_stats_1.csv",header=None)
df1.columns= ['sRNA','count']
df2 = pd.read_csv("model_stats_2.csv",header=None)
df2.columns= ['sRNA','count']
df3 = pd.read_csv("model_stats_3.csv",header=None)
df3.columns= ['sRNA','count']
df4 = pd.read_csv("model_stats_4.csv",header=None)
df4.columns= ['sRNA','count']
df5 = pd.read_csv("model_stats_5.csv",header=None)
df5.columns= ['sRNA','count']
df6 = pd.read_csv("model_stats_6.csv",header=None)
df6.columns= ['sRNA','count']

plot = sns.catplot(x='count',y='sRNA',data=df1, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot1.svg",format="svg")

plot = sns.catplot(x='count',y='sRNA',data=df2, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot2.svg",format="svg")

plot = sns.catplot(x='count',y='sRNA',data=df3, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot3.svg",format="svg")

plot = sns.catplot(x='count',y='sRNA',data=df4, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot4.svg",format="svg")

plot = sns.catplot(x='count',y='sRNA',data=df5, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot5.svg",format="svg")

plot = sns.catplot(x='count',y='sRNA',data=df6, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plot.set(xlim=(0,350))
plt.savefig("plot6.svg",format="svg")

```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("model_stats.csv",header=None)
df.columns=['sRNA','count']
df.head()
df2 = df[df['count'] > 2]
#print(df2)
plot = sns.catplot(x='count',y='sRNA',data=df2, height=10,aspect=0.3,jitter=False,sharex=False, sharey=False)
plt.title("sRNA annotations per genome")
plt.xlabel("Copy-number per genome")
plt.ylabel("HMM")
plot.set(xlim=(0,350))
plt.savefig("plot.svg",format="svg")
```
