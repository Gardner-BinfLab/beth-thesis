```python
%%bash
echo 'sRNA,Total_annotations,PID_change' > turnover_vs_inheritance.csv
for x in {2..281}; do
    name=$( head -n 1 Salmonella.genus.identity | cut -d',' -f$x )
    echo -n ${name},
    cut -d',' -f$x *.genus.identity |\
    grep -v ${name} |\
    awk 'BEGIN{ a=100 } $1~/^[0-9]+$/ {sum+=1; if ($a > $1) ; a = $1 } END{print sum","(100-a)}'
done >> turnover_vs_inheritance.csv
```

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

turnover = pd.read_csv("turnover_vs_inheritance.csv")
inheritance= pd.read_csv("inheritance_NZMS.csv")
inheritance.columns=['sRNA','Inheritance type']
df = pd.merge(turnover,inheritance,on='sRNA')

sns.scatterplot(x='PID_change',y='Total_annotations',hue='Inheritance type',data=df)
plt.show()
```

```python
import numpy as np

exp = pd.read_csv("hinton_relative_expression.csv", header=None)
exp['max'] = exp.max(axis=1)
exp['max'] = np.log(exp['max'])
exp.rename(columns = {list(exp)[0]: 'sRNA'}, inplace = True)
df2 = pd.merge(exp,df,on='sRNA')
df2.head()
plot=sns.scatterplot(x='max',y='Total_annotations',hue='Inheritance type',data=df2)
#plot.set_xlim(auto=True)
#plot.set(xscale='log')
plt.show()


```

```python
df = pd.read_csv("all.csv")
inheritance= pd.read_csv("inheritance_2.tsv",sep="\t")
inheritance.columns=['sRNA','Inheritance type']
df = pd.merge(df,inheritance,on='sRNA')
df['Ratio'] = df['Total_annotations']/(df['Total_annotations']+df['Missing_in'])

plot = sns.boxplot(x='Inheritance type',y='Ratio',data=df)
plot = sns.swarmplot(x='Inheritance type',y='Ratio',data=df,color='.25')
plot.set_xticklabels(rotation=0)
plt.show()
```
