```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("backofen_diff_start_finish.txt",sep="\t")
df.columns=['start','finish']
df.head()
%matplotlib inline
plt.figure(figsize=(11, 7))
sns.set_context("paper", font_scale=2)
sns.distplot( df["start"] , color="skyblue", kde=False,label="Start")
sns.distplot( df["finish"] ,  color="red", kde=False, label="Finish")

plt.legend(prop={'size': 12})
plt.xlabel('Distance from known annotation (nt)')
plt.ylabel('Density')

plt.show()
#plt.savefig("start_finish.svg",format="svg")
```
